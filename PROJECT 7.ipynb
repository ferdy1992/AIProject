```python
languages = ['c++','java','switf','objective -c','Ruby']
popularity = [40,60,10,15,70]
```

```python
plt.bar(languages, popularity)
plt.show()
```

```python
plt.bar(languages, popularity)
plt.title('languages with popularity level')
plt.xlabel('languages')
plt.ylabel('popularity')
plt.show()
```

```python
plt.pie(popularity, labels=languages)
plt.show()
```

```python
plt.pie(popularity, labels = languages, autopct='%1.1f%%')
plt.show()
```

```python
year = [1990, 1991, 1992, 1993, 1994]
rice_revenue = [4500, 5800, 6700, 3200, 4500]
corn_revenue = [7600, 8000, 5500, 2300, 9000]
```

```python
plt.plot(year, rice_revenue, marker = 'x', label = 'Rice')
plt.plot(year, corn_revenue, marker = 'o', label = 'Corn')
plt.legend()
plt.title('revenue from 1990 TO 1994')
plt.xlabel('years')
plt.ylabel('Revenue')
plt.xticks(year)
plt.figure(figsize=(10,6))
plt.show()
```
